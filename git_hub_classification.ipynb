```python
#Importing libraries
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import normalize
from sklearn.model_selection import train_test_split
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.ensemble import RandomForestClassifier, VotingClassifier
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
#Loading the data and pre-processing
Data = pd.read_excel("/content/drive/MyDrive/mk_final.xlsx")
y1 = list(Data['Y1'])
y2 = list(Data['Y2'])
y3 = []
for i,j in enumerate(y1):
  if y2[i]<0.65 and j<-0.03: #Cut of barrier
    y3.append(1)
  else:
    y3.append(0)
Data1 = Data.drop('Metal',axis = 1)
Data2 = Data1.drop('Y1',axis = 1)
Data3 = Data2.drop('Y2',axis = 1)
x3 = Data3
#Models to be used 
rf = RandomForestClassifier()
svm = SVC()
log = LogisticRegression()


for i in range(30):
  x3_train, x3_test, y3_train, y3_test = train_test_split(x3, y3, test_size = 0.25, random_state = i,shuffle=True)
  scaler=StandardScaler()
  X_train = normalize(scaler.fit_transform(x3_train))
  X_test = normalize(scaler.fit_transform(x3_test))
  rf = RandomForestClassifier(n_estimators=10, max_depth = 2, random_state=0)  
  svm = SVC(probability=True, random_state=0,C =3) 
  log = LogisticRegression(C=10)
  
  clf_vot = VotingClassifier(estimators = [('rf',rf), ('svm',svm), ('log', log)], voting='soft') 
  

# training 
  clf_vot.fit(X_train, y3_train)
  rf.fit(X_train, y3_train)
  log.fit(X_train, y3_train)
  svm.fit(X_train, y3_train)

#testing 
  y3_pred_prime = clf_vot.predict_proba(X_test)
  log_pred = log.predict_proba(X_test)
  svm_pred = svm.predict_proba(X_test)
  rf_pred = rf.predict_proba(X_test)
  print(i,x3_test,y3_pred_prime,log_pred,svm_pred,rf_pred)

```
